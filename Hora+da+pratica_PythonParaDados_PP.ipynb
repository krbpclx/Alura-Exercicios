A partir do valor da variação, deve ser enviada às seguintes sugestões:
- Para variação acima de 20%: bonificação para o time de vendas.
- Para variação entre 2% e 20%: pequena bonificação para time de vendas.
- Para variação entre 2% e -10%: planejamento de políticas de incentivo às vendas.

````python
temperatura = float(input('Digite a temperatura em graus Celsius: '))

contador = 0
soma = 0

while temperatura != -273:
  contador += 1
  soma += temperatura
  temperatura = float(input('Digite a temperatura em graus Celsius: '))

media = soma / contador
print(f'A média das temperaturas é: {media}')
````

````python
n = int(input('Digite um número inteiro: '))

fatorial = 1
contador = 1

while contador <= n:
  fatorial *= contador
  contador += 1

print(f'O fatorial de {n} é {fatorial}')
````

Tendo esses valores, faça um código que gere uma lista contendo o percentual de crescimento de bactérias por dia, comparando o número de bactérias em cada dia com o número de bactérias do dia anterior. Dica: para calcular o percentual de crescimento usamos a seguinte equação: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
lista = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]

lista_percentual = []

for i in range(1,len(lista)):
  percentual = 100 * (lista[i] - lista[i-1]) / lista[i-1]
  lista_percentual.append(percentual)
print(f'Porcentagens de crescimento:\n{lista_percentual}')
````
